#nltk.Text
This module brings together a variety of NLTK functionality for text analysis, and provides simple, interactive interfaces. 

Functionality includes: concordancing, collocation discovery, regular expression search over tokenized strings, and distributional similarity.

(source: http://www.nltk.org/api/nltk.text.html)

**concordance(word, width=79, lines=25)**:
Prints a concordance for word with the specified context window. Word matching is not case-sensitive.
* word (str or list) – The target word or phrase (a list of strings)
* width (int) – The width of each line, in characters (default=80)
* lines (int) – The number of lines to display (default=25)

**Thí dụ 3**: Tìm từ 'Stage' xuất hiện trong văn bản text

```python
text = nltk.Text(macbeth)
text.concordance('Stage')
```

**common_contexts(words, num=20)**: Find contexts where the specified words appear; list most frequent common contexts first.
* words (str) – The words used to seed the similarity search
* num (int) – The number of words to generate (default=20)

**Thí dụ 4**: Tìm từ xuất hiện trước và sau từ 'Stage'

```python
text.common_contexts(['Stage'], num= 2)
```

**similar(word, num=20)**: Distributional similarity: find other words which appear in the same contexts as the specified word; list most similar words first.
* word (str) – The word used to seed the similarity search
* num (int) – The number of words to generate (default=20)

**Thí dụ 5**: Tìm từ tương tự từ Stage

```python
text.similar('Stage')
```

**collocations(num=20, window_size=2)**: Print collocations derived from the text, ignoring stopwords

```python
nltk.download('stopwords')
text.collocations(num=10)
```

# Stopword
Nó là những từ thông dụng, thường ít có ý nghĩa trong quá trình phân tích văn bản và thường cần được loại bỏ

```python
nltk.download('stopwords')
```

Muốn xem các stopword trong tiếng Anh, dùng lệnh

```python
sw = nltk.corpus.stopwords.words('english')
print(len(sw))
list(sw)[:10]
```

Có 179 stopword trong từ vựng tiếng Anh. Ta sẽ loại bỏ các từ stopword trong biến macbeth

```python
macbeth_filtered = [w for w in macbeth if w.lower() not in sw]
len(macbeth_filtered)
```

fd = nltk.FreqDist(macbeth_filtered)
fd.most_common(10)

```python
fd = nltk.FreqDist(macbeth_filtered)
fd.most_common(10)
```

Bây giờ, 10 từ phổ biến nhất đầu tiên được trả về, các stopword đã được loại bỏ nhưng vẫn còn các dấu câu, ta cần loại bỏ các dấu câu theo lệnh sau:

```python
import string
punctuation = set(string.punctuation)
macbeth_filtered2 = [w.lower() for w in macbeth if w.lower() not in sw and w.lower() not in punctuation]
fd = nltk.FreqDist(macbeth_filtered2)
fd.most_common(10)
```

# Lựa chọn các từ trong văn bản

**Lựa chọn các từ trong văn bản**:
Một dạng khác trong việc xử lý và phân tích dữ liệu là quá trình chọn một từ dựa vào 1 đặc điểm nào đó. Ví dụ như bạn cần quan tâm rút trích từ dựa vào độ dài của từ. Rút trích các từ có độ dài lớn nhất, ví dụ các từ có độ dài lớn hơn 12 ký tự , dùng lệnh sau:

```python
long_words = [w for w in macbeth if len(w)> 12]
sorted(long_words)
```

Rút trích các từ có chứa chuỗi ‗ious‘


```python
ious_words = [w for w in macbeth if 'ious' in w]
ious_words = set(ious_words)
sorted(ious_words)
```

# Bigrams và collocations

Một phần cơ bản trong phân tích văn bản là xem xét các cặp từ thay thế cho các từ đơn. Từ 'is' và 'yellow' là 1 ví dụ về bigram vì chúng có thể kết hợp lại và có ý nghĩa, vì thế 'is yellow' có thể được tìm thấy trong văn bản. Ví dụ 'fast food', 'pay attention', 'good morning',..,những bigram này gọi là collocation.
Lọc các bigram sau khi đã loại các stopword và các dấu câu, dùng lệnh sau:

```python
bgrms = nltk.FreqDist(nltk.bigrams(macbeth_filtered2))
bgrms.most_common(15)
```

Ngoài bigram ra, còn có trigram, sự kết hợp của 3 từ, ta dùng lệnh trigrams()

```python
tgrms = nltk.FreqDist(nltk.trigrams (macbeth_filtered2))
tgrms.most_common(10)
```

**Tập dữ liệu mẫu**: tập dữ liệu mẫu được chọn từ tập dữ liệu **movie_reviews** được tải từ gói nltk. Thông tin trong tập dữ liệu mẫu này bao gồm, nội dung reviews, ý kiến nhận xét (negative, neutral, positive): 

```python
import random
reviews = nltk.corpus.movie_reviews
documents = [(list(reviews.words(fileid)), category)
for category in reviews.categories()
for fileid in reviews.fileids(category)]
random.shuffle(documents)
```

```python
all_words = nltk.FreqDist(w.lower() for w in reviews.words())
word_features = list(all_words)
```

```python
def document_features(document, word_features):
  document_words = set(document)
  features = {}
  for word in word_features:
    features['{}'.format(word)] = (word in document_words)
  return features
```

```python
featuresets = [(document_features(d,word_features), c) for (d,c) in documents]
len(featuresets)
```

```python
train_set, test_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
train_set, est_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```
